# COLOR SETTINGS

```python
import matplotlib.pyplot as plt
import numpy as np

cm = plt.get_cmap('Oranges')

markers = ["o", "x", "v", "D", "*", "1", ".", "<", ">", "^", "2", "3"]

print(cm)

def update_colors(ax):
    lines = ax.lines
    colors = cm(np.linspace(0.3, 1, len(lines)))
    for line, c in zip(lines, colors):
        line.set_color(c)
```

# TOP-K EASY+HARD

```python
import matplotlib.pyplot as plt
import sys

x = [1,2,3,5,10,20,30,50]
y = {
    "Ours": [64, 70, 73, 77, 83, 89, 92, 96],
    "re-OBJ": [36, 44, 50, 57, 68, 80, 87, 94],
}

fig, ax = plt.subplots(figsize=(10, 10))

for idx, (name, ys) in enumerate(y.items()):
    ax.plot(x, ys, label=name, marker=markers[idx], linestyle=":")

update_colors(ax)
leg = ax.legend(loc="lower right")

fig.suptitle('Top-K Accuracy Easy Negatives', fontsize=20)
plt.xlabel('Top-K', fontsize=18)
plt.ylabel('Accuracy', fontsize=16)
plt.xticks(x)

plt.show()
```
